# P-ALICE, weighted linear regression

```python
X_pool_linear = np.vstack((X.T,np.ones(len(X)))).T
X_pool_linear
```

```python
num_features_linear = len(X_pool_linear[0])
basis_funcs_linear: List[np.ufunc] = [pool_alice_code.identity_basis for _ in range(num_features_linear)] # type: ignore
try:
    Inhibition_data_linear_palice_4: np.ndarray = np.load('../results/Inhibition_data_linear_palice_4.npy')
    print("RESULTS ALREADY OBTAINED")
except:
    Inhibition_data_linear_palice_4 = pool_alice_code.simulate_pool_alice_n_times(n=len(SEEDS),seeds=SEEDS,X_pool=X_pool_linear,y_pool=y,basis_funcs=basis_funcs_linear,batch_size=BATCH_SIZE,passive_learning=False)
    np.save('../results/Inhibition_data_linear_palice_4.npy', Inhibition_data_linear_palice_4)
```
