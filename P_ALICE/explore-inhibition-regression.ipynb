# Load Data

```python
X: np.ndarray = np.load('../data/Inhibition_data/X.npy')
y: np.ndarray  = np.load('../data/Inhibition_data/y.npy')
columns='pr_modifications,pr_production_source_organism,Beta_Sheet,Alpha_Helix,A_Percentage,R_Percentage,N_Percentage,D_Percentage,C_Percentage,Q_Percentage,E_Percentage,G_Percentage,H_Percentage,I_Percentage,L_Percentage,K_Percentage,M_Percentage,F_Percentage,P_Percentage,S_Percentage,T_Percentage,W_Percentage,Y_Percentage,V_Percentage,Isoelectric_Point,Molecular_Weight,Aromaticity,Instability_Index,GRAVY'.split(',')
X_df = pd.DataFrame(X,columns=columns)
X_df = X_df.drop(labels=['pr_production_source_organism'],axis=1)
X: pd.DataFrame = X_df
num_original_features = len(X_df.columns)
```

# SKIPPED: exploring higher order features and nonlinear transformations for nonlinear regression
- already tried out adding some higher order features to this dataset (in addition to the original ones), since was not computationally expensive. the num_samples is too small to double the features by adding X^2 terms. we need to add a subset of additional terms and perhaps transform the original features themselves

```python
num_original_features = len(X.iloc[0]) # first_row = df.iloc[0]
print(num_original_features)

X_extra = copy.deepcopy(X)
```

```python
X_extra.head()
```

```python
for i in range(num_original_features):
    feature_i_name = X_extra.columns[i]

    # add squared feature
    feature_i_name_squared = feature_i_name + '^2'
    X_extra[feature_i_name_squared] = X_extra[feature_i_name]**2

    # add cubed feature
    feature_i_name_cubed = feature_i_name + '^3'
    X_extra[feature_i_name_cubed] = X_extra[feature_i_name]**3

    # add first order interactions
    for j in range(i,num_original_features): # ranging from i to num_original_features prevents X1*X2 and X2*X1
        if i != j:
            feature_j_name = X_extra.columns[j]
            interaction_name = feature_i_name + '*' + feature_j_name
            X_extra[interaction_name] = X_extra[feature_i_name] * X_extra[feature_j_name]
```

```python
X_extra.head()
```

```python
X_extra.shape
```

```python
X_extra['target'] = y
```

```python
X_extra.shape
```

```python
# result is two symmetric matrices, where
# matrix_1[i][j] refers to the spearman coefficient for the ith feature against the jth feature. we are only interested in the last column (which is the transpose of the last row), which is the spearman coefficient between the target and each feature (since we added the target to the dataframe, above)
res = stats.spearmanr(a=X_extra)
```

```python
res = np.array(res)
```

```python
res.shape
```

```python
# Spearman correlation matrix or correlation coefficient (if only 2 variables are given as parameters). Correlation matrix is square with length equal to total number of variables (columns or rows) in a and b combined. i.e. the spearman test statistic 
spearmean_coefs = res[0,:,:] 
# The p-value for a hypothesis test whose null hypothesis is that two samples have no ordinal correlation. See alternative above for alternative hypotheses. pvalue has the same shape as statistic.
p_vals = res[1,:,:]

print(f"spearmean_coefs.shape={spearmean_coefs.shape}, p_vals={p_vals.shape}")
```

```python
def check_symmetric(a, rtol=1e-05, atol=1e-08):
    # https://stackoverflow.com/questions/42908334/checking-if-a-matrix-is-symmetric-in-numpy
    return np.allclose(a, a.T, rtol=rtol, atol=atol)

def check_diagonals_zero(a, rtol=1e-05, atol=1e-08):
    return np.allclose(np.diag(a),0,rtol=rtol,atol=atol)

print(check_symmetric(p_vals))
print(check_symmetric(spearmean_coefs))
if check_diagonals_zero(p_vals) == True:
    print("All features spearman coefficients against themselves checks out wrt to p-val") # all features are exactly correlated with themselves, so we can confidently reject the null hypothesis (see above) 
else:
    print("All features spearman coefficients against themselves DO NOT CHECK OUT wrt to p-val. something is wrong") 

print(check_diagonals_zero(spearmean_coefs))
```

```python
last_row_idx = len(p_vals)-1
p_vals_y_vs_all = p_vals[last_row_idx,:]
print(last_row_idx)
p_vals_y_vs_all.shape
```

```python
p_val_dict = {}

for feature_idx, p_val_y in enumerate(p_vals_y_vs_all):
    feature = X_extra.columns[feature_idx]
    if p_val_y < 0.001:
        p_val_dict[0.001] = p_val_dict.get(0.001,[]) + [(feature, p_val_y)]
    elif p_val_y < 0.01:
        p_val_dict[0.01] = p_val_dict.get(0.01,[]) + [(feature, p_val_y)]
    elif p_val_y < 0.05:
        p_val_dict[0.05] = p_val_dict.get(0.05,[]) + [(feature, p_val_y)]
    else:
        p_val_dict['big'] = p_val_dict.get('big',[]) + [(feature, p_val_y)]
```

```python
p_val_dict
```

```python
for key in p_val_dict:
    print(f"len(p_val_dict[{key}])={len(p_val_dict[key])}")
    print(p_val_dict[key])
```

```python
list_best_features: List[str] = []

for key in p_val_dict:
    if key != "big":
        list_best_features.extend([item[0] for item in p_val_dict[key]])
```

```python
list_best_features.sort()
list_best_features
```

```python
try:
    del list_best_features[list_best_features.index('target')]
except:
    print("already removed target from the significant targets list!")
list_best_features
```

Based on the above, we will try using anything that was below 0.05 in p_val (which is the keys 0.05, 0.001, 0.01)

```python
# X_extra_numpy = X_extra.to_numpy()
# np.save('./X_abalone_pretransformation_1.npy',X_extra_numpy)
```

```python
basis_funcs = []
objs_for_pretransform_columns = []

for feature in list_best_features:
    if "*" in feature:
        feature_original_1, feature_original_2 = feature.split("*")
        # print(feature_original_1, feature_original_2) # verify above!

        # add the column of values to our list (which will be a dataframe series)
        objs_for_pretransform_columns.append(X_df[feature_original_1] * X_df[feature_original_2])
        
        # the two features are multiplied together element wise already, so we just need to find a parameter for this interaction like we would in simple linearn regression
        basis_funcs.append(pool_alice_code.identity_basis) 
        
    elif "^" in feature:
        feature_original, power = feature.split("^")
        # print(feature_original, power) # verify above!

        # add the column of values to our list (which will be a dataframe series)
        objs_for_pretransform_columns.append(X_df[feature_original])
        
        if power == '2':
            basis_funcs.append(pool_alice_code.squared_basis) 
        else:
            basis_funcs.append(pool_alice_code.cubed_basis) 

    else:
        feature_original = feature
        # print(feature_original, 'original') # verify above!

        objs_for_pretransform_columns.append(X_df[feature_original])
        basis_funcs.append(pool_alice_code.identity_basis) 
        
assert len(objs_for_pretransform_columns) == len(basis_funcs) == len(list_best_features), 'messed up!'
```

```python
X_pretransformation_nonlinear_1 = pd.concat(objs_for_pretransform_columns,axis=1)
X_pretransformation_nonlinear_1.columns = list_best_features
X_pretransformation_nonlinear_1.head()
```

```python
X_transformed_1 = pool_alice_code.apply_basis_funcs(X_pretransformation_nonlinear_1.to_numpy(),basis_funcs)

pd.DataFrame(X_transformed_1).head()
```
