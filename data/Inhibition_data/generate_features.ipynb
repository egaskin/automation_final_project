```python
import pandas as pd
from Bio.SeqUtils.ProtParam import ProteinAnalysis
from Bio.SeqUtils import ProtParamData
import requests
from Bio import SeqIO
from io import StringIO
```

```python
metadata = pd.read_csv('Metadata/Protein_Metadata.txt',sep='\t')
metadata.rename(columns={'verf_pr_uniprot_id': 'UniProt_ID'}, inplace=True)
# drop row that does not contain sequence from the data
metadata = metadata[metadata['UniProt_ID'] != 'P0A5S4']
metadata.rename(columns={'verf_pr_uniprot_id': 'UniProt_ID'}, inplace=True)
metadata
```

```python
def add_sequence_column(df, id_column_name='UniProt_ID'):
    def get_protein_sequence(uniprot_id):
  
        baseUrl = "http://www.uniprot.org/uniprot/"
        currentUrl = baseUrl + uniprot_id + ".fasta"
        response = requests.get(currentUrl)
        print(uniprot_id)
        if response.ok:
            Seq = StringIO(response.text)
            pSeq = list(SeqIO.parse(Seq, 'fasta'))

            protein_sequence = str(pSeq[0].seq)
            return protein_sequence
        else:
            print(f"Failed to retrieve data from UniProt for ID: {uniprot_id}")
            return None
    df['Sequence'] = df[id_column_name].apply(get_protein_sequence)
    return df
```

```python
metadata = add_sequence_column(metadata)
metadata
```

```python
# induce the mutations to the protein sequence

def induce_mutation(instruction, sequence):
    print('instruction',instruction)
    if instruction == 'none specified':
        return sequence
    
    mutations = instruction.split(',')
    mutated_sequence = sequence

    for mutation in mutations:
        # deletions
        if '-' in mutation:
            start, end = mutation.split('-')
            print(start,end)
            start = int(start[1:])
            end = int(end[1:-3])
            print(start,end)
            mutated_sequence = mutated_sequence[:start - 1] + mutated_sequence[end:]
        elif 'ins' in mutation:
            inserted_aa = mutation[0]
            mutated_sequence = mutated_sequence[:start-1] + inserted_aa + mutated_sequence[start:]
        else:
            position = int(mutation[1:-1])  # Extract position from instruction
            mutated_aa = mutation[-1]  # Extract mutated amino acid from instruction
            true_aa = mutation[0]
            mutated_sequence = mutated_sequence[:position - 1] + mutated_aa + mutated_sequence[position:]

        return mutated_sequence


# Apply the function to induce mutation
metadata['Sequence'] = metadata.apply(lambda row: induce_mutation(row['pr_mutations'], row['Sequence']), axis=1)

metadata
```

```python
def generate_protein_features(df, sequence_column='Sequence'):
    beta_sheet = []
    alpha_helix = []
    amino_acid_percentages = {aa: [] for aa in ProtParamData.kd.keys()}
    isoelectric_point = []
    molecular_weight = []
    aromaticity = []
    instability_index = []
    gravy = []
    
    for sequence in df[sequence_column]:
        analysis = ProteinAnalysis(sequence)
        beta_sheet.append(analysis.secondary_structure_fraction()[0])
        alpha_helix.append(analysis.secondary_structure_fraction()[1])
        for aa, percent in analysis.get_amino_acids_percent().items():
            amino_acid_percentages[aa].append(percent)
        isoelectric_point.append(analysis.isoelectric_point())
        molecular_weight.append(analysis.molecular_weight())
        aromaticity.append(analysis.aromaticity())
        instability_index.append(analysis.instability_index())
        gravy.append(analysis.gravy())
    
    df['Beta_Sheet'] = beta_sheet
    df['Alpha_Helix'] = alpha_helix
    for aa, percent_list in amino_acid_percentages.items():
        df[f'{aa}_Percentage'] = percent_list
    df['Isoelectric_Point'] = isoelectric_point
    df['Molecular_Weight'] = molecular_weight
    df['Aromaticity'] = aromaticity
    df['Instability_Index'] = instability_index
    df['GRAVY'] = gravy
    
    return df
```

```python
metadata = generate_protein_features(metadata)
metadata
```

```python
def map_modification(modification):
    if modification == 'phosphorylated':
        return 1
    else:
        return 0
    
metadata['pr_modifications'] = metadata['pr_modifications'].apply(lambda x: map_modification(x))

def map_source(source):
    if source == 'Bacterial expression system':
        return 0
    else:
        return 1
metadata['pr_production_source_organism'] = metadata['pr_production_source_organism'].apply(lambda x: map_modification(x))
```
